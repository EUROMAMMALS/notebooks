```python
output = {}
for net in networks:
    dbname = "{net}_db".format(net=net)
    db_connection_url = "postgresql://{us}:{pas}@{host}:{port}/{db}".format(us=dbsett['user'],
                                                                      pas=dbsett['password'],
                                                                      host=dbsett['host'],
                                                                      port=dbsett['port'],
                                                                      db=dbname
                                                                     )
    con = create_engine(db_connection_url)
    connection = con.connect()
    inspector = inspect(con)
    tables = inspector.get_table_names(schema="main")
    for tab in tables:
        meta = MetaData(schema="main")
        meta.reflect(bind=con)
        meta.tables[f"main.{tab}"]
        if tab not in output.keys():
            output[tab] = {}
        output[tab][net] = meta.tables[f"main.{tab}"]
```
